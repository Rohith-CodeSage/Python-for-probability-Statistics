```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/cricket_dataset.csv')
data = df
# Histogram
def plot_histogram(df,column,bins=10):
  plt.figure(figsize=(8, 6))
  sns.histplot(df[column], bins = bins, kde = True, color = 'blue')
  plt.title(f'Histogram of {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()

plot_histogram(data,'Wickets')

# scatter plot
def plot_scatterplot(df, x_column, y_column):
  plt.figure(figsize=(8, 6))
  sns.scatterplot(x=df[x_column], y=df[y_column], color='green')
  plt.title(f'Scatter Plot of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()

plot_scatterplot(data,'Matches', 'Runs')

#bar

def plot_barChart(df, x_column, y_column):
  plt.figure(figsize=(8, 6))
  sns.barplot(x=df[x_column], y=df[y_column], palette='viridis')
  plt.title(f'Bar chart  of {x_column} vs {y_column}')
  plt.xlabel(f'{x_column}')
  plt.xlabel(f'{y_column}')
  plt.show()

plot_barChart(data,'Matches', 'Runs')
```
